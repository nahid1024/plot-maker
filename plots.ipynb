```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator
import random
%matplotlib inline
```

```python
def get_random_rgb_color():
    
    return (random.random() * 0.5, random.random() * 0.5, random.random() * 0.5)
```

```python
def makePlot(x,y, formula, figno):
    # create a larger figure so the plot isn't cramped
    plt.figure(figsize=(10,6))
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title(r"Graph of " + formula, fontsize = 20)
    plt.grid(True, alpha=0.3)
    plt.plot(x, y, color=get_random_rgb_color())
    plt.legend([formula])
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    ax = plt.gca()
    # Force integer (whole-number) major ticks on both axes
    ax.xaxis.set_major_locator(MultipleLocator(1))
    ax.yaxis.set_major_locator(MultipleLocator(1))
    plt.xticks(fontsize = 14)
    plt.yticks(fontsize = 14)
    #plt.xlim(-10, 10)   # x-axis from 0 to 3
    #plt.ylim(-2, 10)
    # improve saved figure spacing and resolution
    plt.tight_layout()
    plt.savefig(figno, bbox_inches='tight', dpi=150)
    plt.show()
```

```python
dom = np.linspace(-5,5,400)
dom = np.array(dom)
```

```python
rang = np.abs(dom)
#rang
```

```python
makePlot(dom, rang, r"$|x|$", "choto_plot.png")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator

x = np.linspace(-5, 5, 400)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Define some functions
functions = [np.abs(x+2), np.abs(x-2), np.abs(x)+2, np.abs(x)-2]

titles = [r"$y = |x+2|$", r"$y = |x-2|$", r"$y = |x|+2$", r"$y = |x|-2$"]

# Flatten axes for easy iteration
axes = axes.flatten()

for ax, y, title in zip(axes, functions, titles):
    ax.plot(x, y, label=title)
    #ax.set_title(f"Graph of {title}")
    ax.grid(True, alpha=0.3)
    ax.legend()
    
    # Set integer ticks for both axes
    ax.xaxis.set_major_locator(MultipleLocator(1))
    ax.yaxis.set_major_locator(MultipleLocator(1))

    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)


fig.text(0.5, 0.5, "Horizontal Shifting", fontsize=12, ha='center')
fig.text(0.5, 0, "Vertical Shifting", fontsize=12, ha='center')

plt.tight_layout()

plt.subplots_adjust(hspace=0.25)
plt.show()
```
